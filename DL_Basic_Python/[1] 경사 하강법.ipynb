## **경사 하강법을 이용한 최소 값 확인**

### 경사 하강법 수식

```python
def gradientDescent(function, x, delta, lr) :
  # 코드를 작성해주세요
  return newX
```

### 시작점

```python
initialX = -1.5
```

### **경사 하강법을 적용하여 최소 값 확인**

```python
newXList, newYList = [], []

startX = initialX
iteration = 10
lr = 1e-1

for i in range(iteration) :
  newX = gradientDescent(function, startX, delta, lr)
  newXList.append(newX)
  newYList.append(function(newX))
  startX = newX
```

## **경사 하강법에 기반한 선형 회귀 분석**

### 정량적 평가를 위한 Scikit-Learn 라이브러리 사용

```python
from sklearn.metrics import r2_score
```

### 선형 모델 함수 샘플링

```python
def function(x) :
  return 10*x + 5
```

### 선형 모델 함수 시각화

```python
x = np.arange(0, 10, 1e-2)
y = function(x)
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Function")
plt.show()
```

### **경사 하강법 기반 매개변수 최신화 함수 (MSE 손실 함수)**

```python
def gradientDescent(input, target, a, b, lr) :
  # 코드를 작성해주세요
  return newA, newB
```

### 시드 고정

```python
np.random.seed(42)
```

### 모델 파라미터 (매개변수) 초기화

```python
a, b = np.random.random(), np.random.random()
print(a, b)
```

### 하이퍼파라이터 선정

```python
iteration, lr = 100, 1e-2
```

### **경사 하강을 통한 파라미터 최신화**

```python
oldA, oldB = a, b

listA, listB = [], []
listLoss = []

for i in range(iteration) :
  pred = oldA * x + b
  target = function(x)

  loss = np.power((pred - target), 2).mean()
  listLoss.append(loss)

  newA, newB = gradientDescent(x, y, oldA, oldB, lr)
  oldA, oldB = newA, newB

  listA.append(oldA)
  listB.append(oldB)
```

### 손실 함수 시각화

```python
plt.figure(figsize=(10, 5))
plt.plot(np.arange(iteration), listLoss)
plt.xlabel("Iteration")
plt.ylabel("MSE Loss")
plt.title("Loss Graph")
plt.show()
```

### **최신화 된 모델 파라미터 출력**

```python
finalA, finalB = listA[-1], listB[-1]
print(finalA, finalB)
```

### **모델 파라미터를 기준으로 결과 추론**

```python
yPred = finalA*x + finalB
```

### **추론 결과의 정성적 및 정략성 평가**

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Ground Truth")
plt.plot(x, yPred, label="Prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Regression Result")
plt.legend(loc="best")
plt.show()
```

```python
r2_score(y, yPred)
```

### **훈련된 모델 파라미터의 외삽법 (Extrapolation) 결과**

```python
x = np.arange(20, 30, 1e-2)
```

```python
y = function(x)
```

```python
yPred = finalA*x + finalB
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Ground Truth")
plt.plot(x, yPred, label="Prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Regression Result")
plt.legend(loc="best")
plt.show()
```

```python
r2_score(y, yPred)
```

## **경사 하강법에 기반한 다중 레이어 모델 기반 회귀 분석**

### 비선형 모델 함수 샘플링

```python
def function(x) :
  return 4*np.power(x, 2) + 2*x + 1
```

```python
def f(x, a, b) :
  return a*x + b

def g(function, x, c, d) :
  return c*x*function + d
```

```python
def gradientDescent(input, target, a, b, c, d, lr) :
  # 코드를 작성해주세요
  return newA, newB, newC, newD
```

### 모델 파라미터 (매개변수) 초기화

```python
a, b, c, d, = np.random.random(), np.random.random(), np.random.random(), np.random.random()
```

### 하이퍼파라이터 선정

```python
iteration, lr = 2000, 1e-7
```

### **경사 하강을 통한 파라미터 최신화**

```python
oldA, oldB, oldC, oldD = a, b, c, d

listA, listB, listC, listD = [], [], [], []
listLoss = []

for i in range(iteration) :
  pred = oldC*x*(oldA*x + oldB) + oldD
  target = function(x)

  loss = np.power((pred - target), 2).mean()
  listLoss.append(loss)

  newA, newB, newC, newD = gradientDescent(x, y, oldA, oldB, oldC, oldD, lr)
  oldA, oldB, oldC, oldD = newA, newB, newC, newD

  listA.append(oldA)
  listB.append(oldB)
  listC.append(oldC)
  listD.append(oldD)
```
